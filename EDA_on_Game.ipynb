### Function for removing outliers

For calculating the outliers in the dataset i want to calculate the quartile deviation for first quartile and and third quartile.Then by subtracting the quartiles i will get my interquartile range.Then i will get the upper bound and lower bound of the specific column.That's why i can identify the outliers in my specific columns

```python

def calculate_outliers(data, column_name):
    #Identify the column that contains the wages data.
    wages_data = data[column_name]

    # Sort the data in ascending order.
    sorted_data = wages_data.sort_values()

    #Calculate the first quartile (Q1) and the third quartile (Q3) of the data.
    Q1 = sorted_data.quantile(0.25)
    Q3 = sorted_data.quantile(0.75)

    #Calculate the interquartile range (IQR).
    IQR = Q3 - Q1

    #Determine the lower and upper bounds for outliers.
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    #Identify outliers
    outliers = sorted_data[(sorted_data < lower_bound) | (sorted_data > upper_bound)]

    return outliers






```

```python
#outliers for Standard Unit Rate $
outlier_1 = calculate_outliers(df,'Wage')
```

```python
import numpy as np
from scipy.stats import norm
df['Potential'] = df['Potential'].astype(int)
df['Wage'] = df['Wage'].astype(int)
df['Weight'] = pd.to_numeric(df['Weight'], errors='coerce')


def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # Using ddof=1 for sample standard deviation
    z_score = norm.ppf((1 + confidence) / 2)

    margin_of_error = z_score * (std_dev / np.sqrt(n))
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound
```

```python
conf_interval_potential = confidence_interval(df['Potential'])
conf_interval_wage = confidence_interval(df['Wage'])
conf_interval_weight = confidence_interval(df['Weight'])
```

#### Proove Central Limit Theorom by using 'potential' column of the game_data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Assuming that df is a DataFrame with a column named 'Potential'
df['Potential'] = df['Potential'].astype(int)

# Rest of the code remains the same
# Function to generate random samples from the 'Potential' column
def generate_samples(data, sample_size, num_samples):
    return [np.random.choice(data, size=sample_size, replace=True) for _ in range(num_samples)]

# Number of samples to generate
num_samples = 1000

# Sample size for each sample
sample_size = 5

# Generate random samples
samples = generate_samples(df['Potential'], sample_size, num_samples)

# Calculate the means of each sample
sample_means = [sample.mean() for sample in samples]

# Plot a histogram of the sample means
plt.hist(sample_means, bins=30, density=True, alpha=0.6, label='Sample Means')

# Overlay a standard normal distribution curve
x = np.linspace(min(sample_means), max(sample_means), 100)
mean_of_sample_means = np.mean(sample_means)
std_dev_of_sample_means = np.std(sample_means, ddof=1)
normal_dist = norm.pdf(x, mean_of_sample_means, std_dev_of_sample_means)
plt.plot(x, normal_dist, 'r', label='Normal Distribution')

plt.xlabel('Sample Means')
plt.ylabel('Density')
plt.title('Central Limit Theorem Demonstration')
plt.legend()
plt.show()
```
